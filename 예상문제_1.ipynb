# 제 2 유형

> 데이터셋
* https://www.kaggle.com/shubh0799/churn-modelling
* 고객의 신상정보 데이터를 통한 회사 서비스 이탈 예측 (종속변수: Exited)
* 평가지표: F1-score

```python
import pandas as pd

train_df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churn/train.csv")
test_df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churn/test.csv")
submission_df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churn/submission.csv")
```

```python
train_df
```

```python
train_df.info()
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

def my_preprocess(df, target="train", encode_list=None):
    df.drop(["RowNumber", "CustomerId", "Surname"], axis=1, inplace=True)
    df = pd.get_dummies(df, columns=["Geography", "Gender"])
    if target == "train":
        # encoder1 = LabelEncoder()
        # encoder2 = LabelEncoder()
        scaler1 = MinMaxScaler()

        # df["Geography"] = encoder1.fit_transform(df["Geography"])
        # df["Gender"] = encoder2.fit_transform(df["Gender"])
        df = scaler1.fit_transform(df)
        # encode_list = [encoder1, encoder2, scaler1]
        encode_list = [scaler1]
        return df, encode_list
    
    elif target == "test":
        # encoder1 = encode_list[0]
        # encoder2 = encode_list[1]
        # scaler1 = encode_list[2]
        scaler1 = encode_list[0]
        
        # df["Geography"] = encoder1.transform(df["Geography"])
        # df["Gender"] = encoder2.transform(df["Gender"])
        df = scaler1.transform(df)

        return df
    
    return None
```

```python
from sklearn.model_selection import train_test_split

X = train_df.drop("Exited", axis=1)
y = train_df["Exited"]
# train, validation split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, stratify=y, random_state=123)
```

```python
# preprocess
X_train_processed, en_list = my_preprocess(X_train, target="train", encode_list=None)
X_valid_processed = my_preprocess(X_valid, target="test", encode_list=en_list)
test_processed = my_preprocess(test_df, target="test", encode_list=en_list)
```

```python
# imbalance data processing
from imblearn.over_sampling import SMOTE
smote_sample = SMOTE(random_state=123)
X_train_smote, y_train_smote = smote_sample.fit_resample(X_train_processed, y_train)
```

```python
corr = pd.DataFrame(X_train_smote).corr()
corr.style.background_gradient()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC

from sklearn.metrics import f1_score
```

* 로지스틱 회귀

```python
model_logit = LogisticRegression(random_state=123)

model_logit.fit(X_train_smote, y_train_smote)
pred = model_logit.predict(X_valid_processed)

f1_score(y_valid, pred)
```

* 랜덤 포레스트

```python
score = []
for depth in range(5, 26):
    model_rf = RandomForestClassifier(max_depth=depth, random_state=123)

    model_rf.fit(X_train_smote, y_train_smote)
    pred = model_rf.predict(X_valid_processed)

    score.append(f1_score(y_valid, pred))

pd.DataFrame(score, index=range(5,26), columns=["f1"]).sort_values("f1")
```

max_depth = 13 일때 가장 f1 스코어가 높음

```python
model_rf = RandomForestClassifier(max_depth=13, random_state=123)
model_rf.fit(X_train_smote, y_train_smote)

print("훈련데이터의 f1 스코어")
print(f1_score(y_train_smote, model_rf.predict(X_train_smote)))

print("검증데이터의 f1 스코어")
print(f1_score(y_valid, model_rf.predict(X_valid_processed)))
```

* 그레디언트 부스팅

```python
score = []
for depth in range(2, 8):
    model_boost = GradientBoostingClassifier(max_depth=depth, random_state=123)
    model_boost.fit(X_train_smote, y_train_smote)
    pred = model_boost.predict(X_valid_processed)
    score.append(f1_score(y_valid, pred))

pd.DataFrame(score, index=range(2,8), columns=["f1"]).sort_values("f1")
```

max_depth가 6일때 f1 스코어가 가장 높음

```python
model_boost = GradientBoostingClassifier(max_depth=6, random_state=123)
model_boost.fit(X_train_smote, y_train_smote)

print("훈련데이터의 f1 스코어")
print(f1_score(y_train_smote, model_boost.predict(X_train_smote)))
print("검증데이터의 f1 스코어")
print(f1_score(y_valid, model_boost.predict(X_valid_processed)))
```

* 서포트 벡터 머신

```python
model_svm = SVC(kernel="rbf", C=0.1, random_state=123)
model_svm.fit(X_train_smote, y_train_smote)

print("훈련데이터 f1 스코어")
print(f1_score(y_train_smote, model_svm.predict(X_train_smote)))
print("검증데이터 f1 스코어")
print(f1_score(y_valid, model_svm.predict(X_valid_processed)))
```

**그레디언트 부스팅의 성능이 가장 우수**

```python
test_pred = model_boost.predict(test_processed)
submission_df["0"] = test_pred
submission_df.to_csv("submission_model.csv", index=False)
```
