```python
import pandas as pd

train_set = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/train.csv")
test_set = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/test.csv")
```

## 전처리함수 작성

```python
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
def my_preprocess(org_df, target="train", encoders=None):
    df = org_df.copy()
    df.drop("id", axis=1, inplace=True)
    df["Arrival Delay in Minutes"] = df["Arrival Delay in Minutes"].fillna(df["Arrival Delay in Minutes"].mean())
    df.reset_index(drop=True, inplace=True)
    
    if target == "train":
        enc1 = OneHotEncoder(sparse_output=False)
        enc2 = OneHotEncoder(sparse_output=False)
        enc3 = OneHotEncoder(sparse_output=False)
        enc4 = OneHotEncoder(sparse_output=False)
        sc1 = MinMaxScaler()
        df = pd.concat([df.drop(columns=["Gender"]), pd.DataFrame(enc1.fit_transform(df[["Gender"]]), columns=[str(col) for col in enc1.categories_[0]])], axis=1)
        df = pd.concat([df.drop(columns=["Customer Type"]), pd.DataFrame(enc2.fit_transform(df[["Customer Type"]]), columns=[str(col) for col in enc2.categories_[0]])], axis=1)
        df = pd.concat([df.drop(columns=["Type of Travel"]), pd.DataFrame(enc3.fit_transform(df[["Type of Travel"]]), columns=[str(col) for col in enc3.categories_[0]])], axis=1)
        df = pd.concat([df.drop(columns=["Class"]), pd.DataFrame(enc4.fit_transform(df[["Class"]]), columns=[str(col) for col in enc4.categories_[0]])], axis=1)
        df = sc1.fit_transform(df)
        return df, [enc1, enc2, enc3, enc4, sc1]
    if target == "test":
        enc1 = encoders[0]
        enc2 = encoders[1]
        enc3 = encoders[2]
        enc4 = encoders[3]
        sc1 = encoders[4]
        df = pd.concat([df.drop(columns=["Gender"]), pd.DataFrame(enc1.transform(df[["Gender"]]), columns=[str(col) for col in enc1.categories_[0]])], axis=1)
        df = pd.concat([df.drop(columns=["Customer Type"]), pd.DataFrame(enc2.transform(df[["Customer Type"]]), columns=[str(col) for col in enc2.categories_[0]])], axis=1)
        df = pd.concat([df.drop(columns=["Type of Travel"]), pd.DataFrame(enc3.transform(df[["Type of Travel"]]), columns=[str(col) for col in enc3.categories_[0]])], axis=1)
        df = pd.concat([df.drop(columns=["Class"]), pd.DataFrame(enc4.transform(df[["Class"]]), columns=[str(col) for col in enc4.categories_[0]])], axis=1)
        df = sc1.transform(df)
        return df
```

## 검증, 훈련 분할

```python
from sklearn.model_selection import train_test_split

X = train_set.drop("satisfaction", axis=1)
y = train_set["satisfaction"].replace({"neutral or dissatisfied": 0, "satisfied": 1})

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, stratify=y, random_state=7)
```

## 전처리 시행

```python
X_train_proc, encoder_lst = my_preprocess(X_train, target="train")
X_val_proc = my_preprocess(X_val, target="test", encoders=encoder_lst)
test_proc = my_preprocess(test_set, target="test", encoders=encoder_lst)
```

## 분석

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
import tensorflow as tf

from sklearn.metrics import accuracy_score, confusion_matrix
```

### 로지스틱 회귀

```python
model_lr = LogisticRegression(max_iter=500)
model_lr.fit(X_train_proc, y_train)
pred = model_lr.predict(X_val_proc)

accuracy_score(y_val, pred)
```

### 랜덤 포레스트

```python
scores = []
for i in range(5, 21):
    model_rf = RandomForestClassifier(max_depth=i, random_state=7)
    model_rf.fit(X_train_proc, y_train)
    pred = model_rf.predict(X_val_proc)
    scores.append(accuracy_score(y_val, pred))
pd.DataFrame(scores, index=range(5, 21), columns=["acc"]).sort_values("acc", ascending=False)
```

```python
model_rf = RandomForestClassifier(max_depth=19, random_state=7)
model_rf.fit(X_train_proc, y_train)
pred = model_rf.predict(X_val_proc)
accuracy_score(y_val, pred)
```

```python
confusion_matrix(y_val, pred)
```

### AdaBoost

```python
model_ab = AdaBoostClassifier(learning_rate=0.3, random_state=7)
model_ab.fit(X_train_proc, y_train)
pred = model_ab.predict(X_val_proc)
accuracy_score(y_val, pred)
```

```python
confusion_matrix(y_val, pred)
```

### XGBoost

```python
scores = []
for i in range(2, 11):
    model_xgb = XGBClassifier(max_depth=i, random_state=7)
    model_xgb.fit(X_train_proc, y_train)
    pred = model_xgb.predict(X_val_proc)
    scores.append(accuracy_score(y_val, pred))
pd.DataFrame(scores, index=range(2, 11), columns=["acc"]).sort_values("acc", ascending=False)
```

```python
model_xgb = XGBClassifier(max_depth=6, random_state=7)
model_xgb.fit(X_train_proc, y_train)
pred = model_xgb.predict(X_val_proc)
accuracy_score(y_val, pred)
```

```python
confusion_matrix(y_val, pred)
```

### KNN

```python
scores = []
for i in range(1, 35, 2):
    model_knn = KNeighborsClassifier(n_neighbors=i)
    model_knn.fit(X_train_proc, y_train)
    pred = model_knn.predict(X_val_proc)
    scores.append(accuracy_score(y_val, pred))
pd.DataFrame(scores, index=range(1, 35, 2), columns=["acc"]).sort_values("acc", ascending=False)
```

```python
model_knn = KNeighborsClassifier(n_neighbors=9)
model_knn.fit(X_train_proc, y_train)
pred = model_knn.predict(X_val_proc)
accuracy_score(y_val, pred)
```

### SVM

```python
model_svm = SVC(C=0.1, kernel="rbf")
model_svm.fit(X_train_proc, y_train)
pred = model_svm.predict(X_val_proc)
accuracy_score(y_val, pred)
```

### ANN

```python
model_ann = tf.keras.Sequential()
model_ann.add(tf.keras.layers.Dense(64, activation="relu"))
model_ann.add(tf.keras.layers.Dropout(0.2))
model_ann.add(tf.keras.layers.Dense(64, activation="relu"))
model_ann.add(tf.keras.layers.Dropout(0.2))
model_ann.add(tf.keras.layers.Dense(64, activation="relu"))
model_ann.add(tf.keras.layers.Dropout(0.2))
model_ann.add(tf.keras.layers.Dense(1, activation="sigmoid"))

model_ann.compile(optimizer="adam", loss=tf.keras.losses.BinaryCrossentropy(), metrics=["accuracy"])

model_ann.fit(X_train_proc, y_train, batch_size=32, epochs=150, validation_split=0.2, callbacks=tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=10))
```

```python
pred = model_ann.predict(X_val_proc)
y_pred = pred.round()
```

```python
accuracy_score(y_val, y_pred)
```
