# 제 2 유형

> 데이터
* https://www.kaggle.com/iabhishekofficial/mobile-price-classification?select=train.csv
* 핸드폰 가격 예측(price_range: 0(저렴) ~ 3(매우비쌈) 범위)
* 평가지표: accuracy

```python
import pandas as pd

train_df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/mobile/train.csv")
test_df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/mobile/test.csv")
```

## 데이터 탐색

```python
train_df.info()
```

```python
test_df.info()
```

```python
train_df.describe()
```

## 검증, 훈련 데이터 분할 및 전처리

```python
test_df = test_df.drop("id", axis=1)
X = train_df.drop("price_range", axis=1)
y = train_df["price_range"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.2, stratify=y, random_state=6)

sc = MinMaxScaler()
X_train = sc.fit_transform(X_train)
X_val = sc.transform(X_val)
test_set = sc.transform(test_df)
```

## 분석

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
import tensorflow as tf

from sklearn.metrics import accuracy_score, confusion_matrix
```

### 로지스틱 회귀

```python
acc_scores = []
```

```python
model_lr = LogisticRegression()
model_lr.fit(X_train, y_train)
pred = model_lr.predict(X_val)

acc_scores.append(accuracy_score(y_val, pred))
acc_scores[0]
```

### 랜덤포레스트

```python
scores = []
for i in range(2, 21):
    model_rf = RandomForestClassifier(max_depth=i, random_state=6)
    model_rf.fit(X_train, y_train)
    pred = model_rf.predict(X_val)
    scores.append(accuracy_score(y_val, pred))
pd.DataFrame(scores, index=range(2, 21), columns=["acc"]).sort_values("acc", ascending=False)
```

```python
model_rf = RandomForestClassifier(max_depth=14, random_state=6)
model_rf.fit(X_train, y_train)
pred = model_rf.predict(X_val)

acc_scores.append(accuracy_score(y_val, pred))
acc_scores[1]
```

### AdaBoost

```python
for i in [0.001, 0.01, 0.05, 0.1, 0.15, 0.2]:
    model_ab = AdaBoostClassifier(learning_rate=i, random_state=6)
    model_ab.fit(X_train, y_train)
    pred = model_ab.predict(X_val)
    print(accuracy_score(y_val, pred))
```

```python
model_ab = AdaBoostClassifier(learning_rate=0.01, random_state=6)
model_ab.fit(X_train, y_train)
pred = model_ab.predict(X_val)
acc_scores.append(accuracy_score(y_val, pred))
acc_scores[2]
```

### XGBoost

```python
scores = []
for i in range(2, 16):
    model_xgb = XGBClassifier(max_depth=i, random_state=6)
    model_xgb.fit(X_train, y_train)
    pred = model_xgb.predict(X_val)
    scores.append(accuracy_score(y_val, pred))
pd.DataFrame(scores, index=range(2, 16), columns=["acc"]).sort_values("acc", ascending=False)
```

```python
model_xgb = XGBClassifier(max_depth=4, random_state=6)
model_xgb.fit(X_train, y_train)
pred = model_xgb.predict(X_val)
acc_scores.append(accuracy_score(y_val, pred))
acc_scores[3]
```

### KNN

```python
scores = []
for i in range(1, 100, 2):
    model_knn = KNeighborsClassifier(n_neighbors=i)
    model_knn.fit(X_train, y_train)
    pred = model_knn.predict(X_val)
    scores.append(accuracy_score(y_val, pred))
pd.DataFrame(scores, index=range(1, 100, 2), columns=["acc"]).sort_values("acc", ascending=False)
```

```python
model_knn = KNeighborsClassifier(n_neighbors=95,)
model_knn.fit(X_train, y_train)
pred = model_knn.predict(X_val)
acc_scores.append(accuracy_score(y_val, pred))
acc_scores[4]
```

### SVM

```python
model_svm = SVC(C=0.21, kernel="linear")
model_svm.fit(X_train, y_train)
pred = model_svm.predict(X_val)

acc_scores.append(accuracy_score(y_val, pred))
acc_scores[5]
```

### ANN

```python
model_ann = tf.keras.Sequential()
model_ann.add(tf.keras.layers.Dense(64, activation="relu"))
model_ann.add(tf.keras.layers.Dropout(0.2))
model_ann.add(tf.keras.layers.Dense(64, activation="relu"))
model_ann.add(tf.keras.layers.Dropout(0.2))
model_ann.add(tf.keras.layers.Dense(64, activation="relu"))
model_ann.add(tf.keras.layers.Dropout(0.2))
model_ann.add(tf.keras.layers.Dense(4, activation="softmax"))

model_ann.compile(optimizer="adam", loss=tf.keras.losses.SparseCategoricalCrossentropy(), metrics=["accuracy"])
```

```python
model_ann.fit(X_train, y_train, batch_size=32, epochs=200, callbacks=tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=10), validation_split=0.2)
```

```python
pred = model_ann.predict(X_val)
```

```python
y_pred = []
for row in pred:
    y_pred.append(list(row).index(max(row)))
```

```python
acc_scores.append(accuracy_score(y_val, y_pred))
acc_scores[6]
```

## 결과 확인

```python
pd.DataFrame(acc_scores, index=["LogisticReg", "RandomForest", "AdaBoost", "XGBoost", "KNN", "SVM", "ANN"])
```
